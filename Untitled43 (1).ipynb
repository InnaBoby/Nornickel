```python
data = data.drop(labels=0)
```

```python
import re


def choose_FM_from_data(n):
    pattern=re.compile(f'\D+[{n}]\..+?')
    pattern_2 = re.compile(f'\D+_{n}')
    FM = []

    for col in data.columns.to_list():
      if re.findall(pattern, col) or re.findall(pattern_2, col):
        FM.append(col)

    return FM
```

```python
train = pd.DataFrame(data['MEAS_DT'])
target_cols = test.drop(columns=['MEAS_DT']).columns.to_list()

for i in tqdm(range(1, 7)):
    #extrtact data for i floatmachine
    FM = choose_FM_from_data(i)
    FM.append('MEAS_DT')
    FM = data[FM]

    #split data for train and target
    targets= []
    for col in FM.columns.to_list():
        if col in target_cols:
           targets.append(col)
    y=data[targets]
    train = train.merge(FM[list(set(FM.columns.to_list()) - set(target_cols))], on = 'MEAS_DT')

    # fit model for each target
    model = CatBoostRegressor()
    params_for_search = {'iterations' : [50, 100, 300],
                        'depth' : [ 3, 5, 8],
                        'learning_rate' : [0.0001, 0.001, 0.5]
                        }

    grid_search = GridSearchCV(estimator=model,
                              param_grid=params_for_search,
                              cv=5,
                              scoring='neg_root_mean_squared_error'
                              )


    scaler = StandardScaler()
    X = scaler.fit_transform(train.drop(columns=['MEAS_DT']))

    for col in targets:
          grid_search.fit(X, y[col])
          best_model = grid_search.best_estimator_
          preds = best_model.predict(X)
          preds = [round(num, 1) for num in preds]
          preds_df = pd.DataFrame({'MEAS_DT': train['MEAS_DT'], f'{col}': preds}).set_index('MEAS_DT').loc[test.set_index('MEAS_DT').index.to_list()]
          test[col] = preds_df[col]
```
